```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Global constants
EPOCHS = 20000
LEARNING_RATE = 0.01
TEST_SIZE = 0.1
CONNECT_DOTS = True

input_size = 4
hidden_size = 20
output_size = 3

class MLP:
    def __init__(self, input_size, hidden_size, output_size, activation='sigmoid'):
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.activation = activation

        # Xavier initialization
        self.W_input_hidden = np.random.randn(self.input_size, self.hidden_size) * np.sqrt(2.0 / (self.input_size + self.hidden_size))
        self.b_input_hidden = np.zeros((1, self.hidden_size))
        self.W_hidden_output = np.random.randn(self.hidden_size, self.output_size) * np.sqrt(2.0 / (self.hidden_size + self.output_size))
        self.b_hidden_output = np.zeros((1, self.output_size))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-np.clip(x, -500, 500)))

    def d_sigmoid(self, x):
        return x * (1 - x)

    def relu(self, x):
        return np.maximum(0, x)

    def d_relu(self, x):
        return np.where(x > 0, 1, 0)

    def tanh(self, x):
        return np.tanh(x)

    def d_tanh(self, x):
        return 1 - x ** 2

    def forward(self, input_data):
        hidden_layer_input = np.dot(input_data, self.W_input_hidden) + self.b_input_hidden
        hidden_layer_output = self.sigmoid(hidden_layer_input)
        output_layer_input = np.dot(hidden_layer_output, self.W_hidden_output) + self.b_hidden_output
        
        if self.activation == 'sigmoid':
            output = self.sigmoid(output_layer_input)
        elif self.activation == 'relu':
            output = self.relu(output_layer_input)
        elif self.activation == 'tanh':
            output = self.tanh(output_layer_input)
        else:
            output = output_layer_input  # Linear
        return hidden_layer_output, output

    def backward(self, input_data, target, hidden_output, output, lr=LEARNING_RATE):
        output_error = target - output
        if self.activation == 'sigmoid':
            output_grad = output_error * self.d_sigmoid(output)
        elif self.activation == 'relu':
            output_grad = output_error * self.d_relu(output)
        elif self.activation == 'tanh':
            output_grad = output_error * self.d_tanh(output)
        else:
            output_grad = output_error  # Linear

        hidden_error = np.dot(output_grad, self.W_hidden_output.T)
        hidden_grad = hidden_error * self.d_sigmoid(hidden_output)

        # Update weights and biases
        self.W_hidden_output += np.dot(hidden_output.T, output_grad) * lr
        self.b_hidden_output += np.sum(output_grad, axis=0, keepdims=True) * lr
        self.W_input_hidden += np.dot(input_data.T, hidden_grad) * lr
        self.b_input_hidden += np.sum(hidden_grad, axis=0, keepdims=True) * lr

    def train(self, input_data, target, X_test, y_test, epochs=EPOCHS, lr=LEARNING_RATE):
        train_losses = []
        test_losses = []
        for epoch in range(epochs):
            hidden_output, output = self.forward(input_data)
            self.backward(input_data, target, hidden_output, output, lr)
            if epoch % 100 == 0:
                train_loss = np.mean((target - output) ** 2)
                _, test_pred = self.forward(X_test)
                test_loss = np.mean((y_test - test_pred) ** 2)
                train_losses.append(train_loss)
                test_losses.append(test_loss)
                print(f"Epoch {epoch}, Train Loss: {train_loss:.6f}, Test Loss: {test_loss:.6f}")
        return train_losses, test_losses

    def evaluate(self, input_data, target):
        _, output = self.forward(input_data)
        loss = np.mean((target - output) ** 2)
        per_output_loss = [np.mean((target[:, i] - output[:, i]) ** 2) for i in range(target.shape[1])]
        sum_os_res = np.sum((target - output) ** 2)
        target_mean = np.mean(target, axis=0)
        sum_os_tot = np.sum((target - target_mean) ** 2)
        r2_aggregated = 1 - (sum_os_res / (sum_os_tot + 1e-8))
        r2_per_output = []
        for i in range(target.shape[1]):
            ss_res = np.sum((target[:, i] - output[:, i]) ** 2)
            ss_tot = np.sum((target[:, i] - np.mean(target[:, i])) ** 2)
            r2_i = 1 - (ss_res / (ss_tot + 1e-8))
            r2_per_output.append(r2_i)
        return loss, per_output_loss, r2_aggregated, r2_per_output, output

def plot_predictions(y_true, y_pred, sample_names, title_prefix=""):
    output_names = ['Flame Width (pixels)', 'Flame Length (pixels)', 'Flame Area (pixel²)']
    for i in range(y_true.shape[1]):
        plt.figure(figsize=(10, 6))
        if CONNECT_DOTS:
            plt.plot(y_true[:, i], 'bo-', label='Actual', markersize=5)
            plt.plot(y_pred[:, i], 'ro-', label='Predicted', markersize=5)
        else:
            plt.scatter(range(len(y_true)), y_true[:, i], color='blue', label='Actual', s=50)
            plt.scatter(range(len(y_pred)), y_pred[:, i], color='red', label='Predicted', s=50)
        for j in range(len(y_true)):
            plt.text(j, y_true[j, i], f'{sample_names[j]}', fontsize=8, ha='right', va='bottom', color='blue')
            plt.text(j, y_pred[j, i], f'{sample_names[j]}', fontsize=8, ha='left', va='top', color='red')
        plt.title(f"{title_prefix} {output_names[i]}: Actual vs Predicted")
        plt.xlabel('Sample Index')
        plt.ylabel(output_names[i])
        plt.legend()
        plt.grid(True)
        plt.show()

def plot_loss_curves(train_losses, test_losses, title_prefix=""):
    plt.figure(figsize=(10, 6))
    plt.plot(range(0, EPOCHS, 100), train_losses, label='Train Loss', color='blue')
    plt.plot(range(0, EPOCHS, 100), test_losses, label='Test Loss', color='red')
    plt.title(f"{title_prefix} Train vs Test Loss")
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()

# Load data
df = pd.read_excel("Parameters_AveragedLengthWidthArea_Modified.xlsx")
X = df[['Power', 'Gas', 'Tangential', 'Axial']].to_numpy()
y = df[['Flame Width AveragePixel', 'Flame Length Average Pixel', 'Flame Area Average Pixel^2']].to_numpy()
sample_names = df.iloc[:, 0].to_numpy()

# Split data for interpolation
X_train, X_test, y_train, y_test, sample_names_train, sample_names_test = train_test_split(
    X, y, sample_names, test_size=TEST_SIZE, random_state=42, shuffle=True
)

# Normalize data
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
y_train_scaled = scaler_y.fit_transform(y_train)
X_test_scaled = scaler_X.transform(X_test)
y_test_scaled = scaler_y.transform(y_test)

# Activation functions to compare
activations = ['sigmoid', 'relu', 'tanh', 'linear']
results = []

for activation in activations:
    print(f"\nTraining with {activation} activation")
    model = MLP(input_size, hidden_size, output_size, activation=activation)
    train_losses, test_losses = model.train(X_train_scaled, y_train_scaled, X_test_scaled, y_test_scaled)

    # Evaluate
    train_loss, train_per_output_loss, train_r2, train_r2_per_output, train_pred = model.evaluate(X_train_scaled, y_train_scaled)
    test_loss, test_per_output_loss, test_r2, test_r2_per_output, test_pred = model.evaluate(X_test_scaled, y_test_scaled)

    results.append({
        'activation': activation,
        'test_r2_avg': np.mean(test_r2_per_output)
    })

    print(f"\n{activation.capitalize()} Activation Results:")
    print(f"Train Total Loss: {train_loss:.6f}")
    print(f"Train Loss per output: {[f'{loss:.6f}' for loss in train_per_output_loss]}")
    print(f"Train R² per output: {train_r2_per_output}")
    print(f"Train Average R²: {np.mean(train_r2_per_output):.6f}")
    print(f"Test Total Loss: {test_loss:.6f}")
    print(f"Test Loss per output: {[f'{loss:.6f}' for loss in test_per_output_loss]}")
    print(f"Test R² per output: {test_r2_per_output}")
    print(f"Test Average R²: {np.mean(test_r2_per_output):.6f}")

    plot_loss_curves(train_losses, test_losses, title_prefix=f"{activation.capitalize()} Activation")
    plot_predictions(y_test_scaled, test_pred, sample_names_test, title_prefix=f"Test Set ({activation.capitalize()})")

# Find best activation function based on average test R²
best_activation = max(results, key=lambda x: x['test_r2_avg'])['activation']
print(f"\nBest activation function for interpolation: {best_activation.capitalize()}")

# Extrapolation
print("\nExtrapolation (Power 1-8 for training, 8-10 for testing)")
df_train = df[df['Power'] <= 8]
df_test = df[df['Power'] > 8]
X_train_ext = df_train[['Power', 'Gas', 'Tangential', 'Axial']].to_numpy()
y_train_ext = df_train[['Flame Width AveragePixel', 'Flame Length Average Pixel', 'Flame Area Average Pixel^2']].to_numpy()
X_test_ext = df_test[['Power', 'Gas', 'Tangential', 'Axial']].to_numpy()
y_test_ext = df_test[['Flame Width AveragePixel', 'Flame Length Average Pixel', 'Flame Area Average Pixel^2']].to_numpy()
sample_names_train_ext = df_train.iloc[:, 0].to_numpy()
sample_names_test_ext = df_test.iloc[:, 0].to_numpy()

# Normalize extrapolation data
X_train_ext_scaled = scaler_X.fit_transform(X_train_ext)
y_train_ext_scaled = scaler_y.fit_transform(y_train_ext)
X_test_ext_scaled = scaler_X.transform(X_test_ext)
y_test_ext_scaled = scaler_y.transform(y_test_ext)

results_ext = []
for activation in activations:
    print(f"\nExtrapolation with {activation} activation")
    model = MLP(input_size, hidden_size, output_size, activation=activation)
    train_losses, test_losses = model.train(X_train_ext_scaled, y_train_ext_scaled, X_test_ext_scaled, y_test_ext_scaled)

    # Evaluate
    train_loss, train_per_output_loss, train_r2, train_r2_per_output, train_pred = model.evaluate(X_train_ext_scaled, y_train_ext_scaled)
    test_loss, test_per_output_loss, test_r2, test_r2_per_output, test_pred = model.evaluate(X_test_ext_scaled, y_test_ext_scaled)

    results_ext.append({
        'activation': activation,
        'test_r2_avg': np.mean(test_r2_per_output)
    })

    print(f"\n{activation.capitalize()} Activation Extrapolation Results:")
    print(f"Train Total Loss: {train_loss:.6f}")
    print(f"Train Loss per output: {[f'{loss:.6f}' for loss in train_per_output_loss]}")
    print(f"Train R² per output: {train_r2_per_output}")
    print(f"Train Average R²: {np.mean(train_r2_per_output):.6f}")
    print(f"Test Total Loss: {test_loss:.6f}")
    print(f"Test Loss per output: {[f'{loss:.6f}' for loss in test_per_output_loss]}")
    print(f"Test R² per output: {test_r2_per_output}")
    print(f"Test Average R²: {np.mean(test_r2_per_output):.6f}")

    plot_loss_curves(train_losses, test_losses, title_prefix=f"Extrapolation {activation.capitalize()} Activation")
    plot_predictions(y_test_ext_scaled, test_pred, sample_names_test_ext, title_prefix=f"Extrapolation Test Set ({activation.capitalize()})")

# Find best activation function for extrapolation
best_activation_ext = max(results_ext, key=lambda x: x['test_r2_avg'])['activation']
print(f"\nBest activation function for extrapolation: {best_activation_ext.capitalize()}")
```
