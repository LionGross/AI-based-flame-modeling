```python
import pandas as pd
import matplotlib.pyplot as plt

# Excel-Datei einlesen (erste Zeile als Überschriften)
df = pd.read_excel('Parameters_AveragedLengthWidthArea_Modified.xlsx', sheet_name='Sheet1', header=0)

# Spalten für Inputs und Outputs
input_cols = [df.columns[i] for i in [2, 4, 5]]  # Spalten C, E, F
output_cols = [df.columns[i] for i in [6, 7, 8]]        # Spalten G, H, I

# 5x3 Subplot-Matrix erstellen
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(9, 9))
fig.suptitle('2D-Streudiagramme: Inputs vs. Outputs', fontsize=16)

# Schleife über Inputs und Outputs
for i, input_col in enumerate(input_cols):
    for j, output_col in enumerate(output_cols):
        # Streudiagramm für aktuelles Input-Output-Paar
        axes[i, j].scatter(df[input_col], df[output_col], color='blue', s=50, alpha=0.6)
        axes[i, j].set_xlabel(input_col)
        axes[i, j].set_ylabel(output_col)
        axes[i, j].grid(True)

# Layout optimieren
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Excel-Datei einlesen (erste Zeile als Überschriften)
df = pd.read_excel('Parameters_AveragedLengthWidthArea_Modified.xlsx', sheet_name='Sheet1', header=0)

# Spalten für Inputs und Outputs
input_cols = [df.columns[i] for i in [2, 4, 5]]  # Spalten C, E, F
output_cols = [df.columns[i] for i in [6, 7, 8]]  # Spalten G, H, I

# Daten nach Identifier (Spalte D) filtern
df_0 = df[df[df.columns[3]] == 0]  # Daten mit Identifier 0
df_1 = df[df[df.columns[3]] == 1]  # Daten mit Identifier 1

# Anzahl der Datenpunkte für jeden Identifier
count_0 = len(df_0)
count_1 = len(df_1)

# Diagramm für Identifier 0
fig0, axes0 = plt.subplots(nrows=3, ncols=3, figsize=(9, 9))
fig0.suptitle(f'2D-Streudiagramme: Inputs vs. Outputs (Identifier 0, {count_0} Datenpunkte)', fontsize=16)

for i, input_col in enumerate(input_cols):
    for j, output_col in enumerate(output_cols):
        axes0[i, j].scatter(df_0[input_col], df_0[output_col], color='blue', s=50, alpha=0.6)
        axes0[i, j].set_xlabel(input_col)
        axes0[i, j].set_ylabel(output_col)
        axes0[i, j].grid(True)

# Layout optimieren
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# Diagramm für Identifier 1
fig1, axes1 = plt.subplots(nrows=3, ncols=3, figsize=(9, 9))
fig1.suptitle(f'2D-Streudiagramme: Inputs vs. Outputs (Identifier 1, {count_1} Datenpunkte)', fontsize=16)

for i, input_col in enumerate(input_cols):
    for j, output_col in enumerate(output_cols):
        axes1[i, j].scatter(df_1[input_col], df_1[output_col], color='red', s=50, alpha=0.6)
        axes1[i, j].set_xlabel(input_col)
        axes1[i, j].set_ylabel(output_col)
        axes1[i, j].grid(True)

# Layout optimieren
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
